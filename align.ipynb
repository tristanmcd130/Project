```python
from datasets import get_dataset_config_names, load_dataset
```

```python
configs = set(get_dataset_config_names("sentence-transformers/parallel-sentences-europarl")) - {"all", "en-et", "en-fi", "en-hu"}
configs
```

```python
sentences = {}
```

```python
for config in configs:
	lang = config.split("-")[1]
	print(lang)
	ds = load_dataset("sentence-transformers/parallel-sentences-europarl", config)
	for row in ds["train"]:
		if row["english"] not in sentences:
			sentences[row["english"]] = {}
		sentences[row["english"]][lang] = row["non_english"]
```

```python
len(sentences)
```

```python
sentences_in_all = {k: v for k, v in sentences.items() if len(v) == len(configs)}
len(sentences_in_all)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(sentences_in_all, orient="index").reset_index().rename(columns={"index": "en"})
df
```
